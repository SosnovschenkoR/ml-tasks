```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
#np.show_config()
%matplotlib inline

x = np.arange(-8, 8, 0.1)
f = 1 / (1 + np.exp(-x))
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
sns.set()
sns.set_style("darkgrid")
x = np.linspace(0, 14, 100) 
#print("A\n", (x))
_ = plt.plot(x, np.sin(x + 1 * .5), label='sin', marker='.', linestyle='none', color='g')
#_ = plt.plot(x, np.sin(x + 3 * .5), marker='.', linestyle='none', color='r')
y_mean = [np.mean(np.sin(x + 1 * .5))] * len(x)
_ = plt.plot(x, y_mean, label='Mean', linestyle='--')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend(loc='upper right')
plt.show()
```

```python
fig = plt.figure()
# Give the figure a title
fig.suptitle("Empty figure")

ax = fig.add_subplot(1,1,1)
# Set the title of plot
ax.set_title("Empty plot")
plt.show()
```
